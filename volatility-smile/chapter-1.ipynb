# 第1章 总览

主要讨论了BSM模型（Black-Scholes-Merton）。BSM已经成为市场上对期权定价的公认模型，尽管这个模型本身并不完美。整本书的主题其实就是对BSM模型的缺陷的探讨，以及对这些缺陷的建模。

BSM的假设：
* 标的物可以连续对冲
* 对冲交易没有额外成本
* 标的物价格服从几何布朗运动
前两个假设比较容易克服，在交易中可以在关键点位调高价格即可。但第三个假设问题很严重。

1987年金融危机前，市场的隐含波动率图像基本是一条直线；危机后，变成了一条曲线。这是因为市场意识到，行权价距离标的价格差相等的看涨期权和看跌期权中，看跌期权更有价值。这意味着期权的价格不能够根据标的物价格直接推到出来，而引入了主观因素。这个因素反映在隐含波动率与行权价的关系上，成为了波动率微笑曲线。

## 对金融工程的讨论

在逻辑连续的世界中，金融工程是在坚实的金融科学基础上，研究如何创造有用的金融工具的学科。其标的物可以是股票、债券等。金融工程所需的数学基础包括微积分、概率论、随机过程、模拟研究及布朗运动。

然而，现实的资产行为并不满足模型描述的假设条件，有时相去甚远。所以，尽管有这么多的数学工具，我们却并没有找到解释金融现象的正确规律，可能永远也找不到。因为没有找到正确的规律，那么用公理化的方法来分析金融就存在问题（公理化用在几何学、物理等比较成功）。数学很重要，但不要过度沉溺在数学中，金融模型中所面临的**数学问题比想象中的少**。所以在分析时，对假设的依赖越少越好，对模型的依赖也是越少越好。当必须使用模型时，越简单的越好。

## 模型

目的
* 根据高流动性资产的价格推出低流动性资产价格
* 证券排序
* 量化直觉
    
风格
* 相对价值模型
* 绝对价值模型

金融业喜欢用相对价值模型对衍生品估价的原因是衍生品更像是由基本原子组合而成的分子。人们通过观察衍生品构成要素的行为方式来判断衍生品的价值。BSM的高明之处在于，可以通过一个股票和债券来构建一个衍生品。
